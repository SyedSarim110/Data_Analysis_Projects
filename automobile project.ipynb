```python
autos.columns = autos.columns.str.replace("date","date ")\
.str.replace("offer","offer ").str.replace("vehicle","vehicle ").str\
.replace("yearOfRegistration","Registration year").str.replace("power","power ")\
.str.replace("monthOfRegistration","Registration month")\
.str.replace("fuel","fuel ").str.replace("notRepairedDamage","unrepair damage")\
.str.replace("date Created","ad Created").str.replace("nrOfPictures","num photos")\
.str.replace("postalCode","postal Code").str.replace("lastSeen","last Seen")

autos.columns = autos.columns.str.replace(" ","_").str.lower()
```

```python
autos.describe()
```

```python
autos.drop(["num_photos","seller","offer_type"],axis=1,inplace =True)
```

```python
autos["price"]
```

```python
autos["price"] = autos["price"].str.replace("$","")
autos["price"]
```

```python
autos["price"] = autos["price"].str.replace(",","").astype(int)
```

```python
# clean "odometer" column by removing bad characters
autos["odometer"] = autos["odometer"].str.replace(",","").str.replace("km","")
```

```python
autos["odometer"] = autos["odometer"].astype(int)
```

```python
autos.columns = autos.columns.str.replace("price","price_$").str.replace("odometer","odometer_km")
```

```python
odometer_freq = {}
for i in autos["odometer_km"]:
    if i not in odometer_freq:
        odometer_freq[i] = 1
    else:
        odometer_freq[i] += 1
odometer_freq
```
