```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
who_data = pd.read_csv("WHO.csv")
```

```python
-who_data.dtypes
```

```python
who_data["Date_reported"]=pd.to_datetime(who_data["Date_reported"])
```

```python
who_data.dtypes
```

```python
who_data.head()
```

```python
pakistan=who_data[who_data["Country"] == "Pakistan"]
```

```python
plt.plot(pakistan["Date_reported"],pakistan["Cumulative_cases"])
plt.title("Pakistan Reported Cases")
plt.xlabel("Date")
plt.ylabel("Reported Cases")
plt.show()
```

```python
def covid(country_name):
    df = who_data[who_data["Country"] == country_name]
    plt.plot(df["Date_reported"],df["Cumulative_cases"])
    plt.xlabel("Date")
    plt.ylabel("Reported Cases")
 #  plt.title("{} Reported Cases".formate())
    plt.show()
    return covid
```

```python
covid("Pakistan")
```

```python
data = who_data["Country"].unique()
```

```python
data[:10]
```

```python
for i in data[:10]:
    covid(i)
```
